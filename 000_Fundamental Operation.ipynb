```python
import torch
import numpy as np
```

```python
numpy_tensor = np.random.randn(10,20)
print(numpy_tensor)
```

```python
pytorch_tensor1 = torch.Tensor(numpy_tensor)
pytorch_tensor2 = torch.from_numpy(numpy_tensor)
print(pytorch_tensor1)
print(pytorch_tensor2)
```

```python
#numpy_array = pytorch_tensor1.numpy()
numpy_array = pytorch_tensor1.cpu().numpy
print(numpy_array)
```

```python
dtype = torch.cuda.FloatTensor
gpu_tensor = torch.randn(10,20).type(dtype)

gpu_tensor = torch.randn(10,20).cuda(0)
print(gpu_tensor)
```

```python
cpu_tensor = gpu_tensor.cpu()
```

```python
print(pytorch_tensor1.shape)
print(pytorch_tensor1.size())
print(pytorch_tensor1.type())
print(pytorch_tensor1.dim())
print(pytorch_tensor1.numel())
```

```python
extensor = torch.randn(3,2).type(torch.float64)
exnum = extensor.numpy()
print(exnum.size)
print(exnum.dtype)
```

```python
x = torch.ones(2,2)
print(x)
print(x.type())
x = x.long()
print(x)
print(x.type())
```

```python
x = torch.randn(4,3)
print(x)
max_value, max_idx = torch.max(x,dim=1)
print(max_value)
print(max_idx)
sum_x = torch.sum(x, dim=1)
print(sum_x)
```

```python
print(x.shape)
x = x.unsqueeze(0)
print(x.shape)

x = x.unsqueeze(1)
print(x.shape)

x = x.squeeze()
print(x.shape)
```

```python
x = torch.randn(3,4,5)
print(x.shape)
print(x)

x = x.permute(1,0,2)
print(x.shape)
print(x)

x = x.transpose(0,2)
print(x.shape)
print(x)
```

```python
x = torch.randn(3,4,5)
print(x.shape)

x = x.view(-1,5)
print(x.shape)

x = x.view(3,20)
print(x.shape)
```

```python
x = torch.randn(3,4)
y = torch.randn(3,4)
z = x + y
print(z)
```

```python
x = torch.ones(3,3)
x.unsqueeze_(0)
print(x.shape)
x.transpose_(1,0)
print(x.shape)
```

```python
x = torch.ones(3,3)
y = torch.ones(3,3)

print(x)
x.add_(y)
print(x)
```

```python
x = torch.ones(4,4).float()
print(x)
x[1:3,1:3] = 2
print(x)
```
